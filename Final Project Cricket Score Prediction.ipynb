```python
total_teams = ['India', 'Sri Lanka', 'New Zealand', 'West Indies', 'Australia',
       'England', 'South Africa', 'Bangladesh', 'Pakistan', 'Ireland',
       'Zimbabwe', 'Afghanistan', 'Netherlands']
```

```python
df = df[(df['battingTeam'].isin(total_teams)) & (df['bowlingTeam'].isin(total_teams))]
```

```python
print(df['battingTeam'].unique())
print(df['bowlingTeam'].unique())
```

```python
df = df[df['delivery_left']<=84]
```

```python
cat_df = pd.get_dummies(data = df, columns = ['battingTeam' , 'bowlingTeam'])
```

```python
cat_df
```

```python
#df['city'].unique()
```

```python
X = cat_df.drop(columns=['Final_Score'])
Y = cat_df['Final_Score']
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
X_test
```

```python
print(X_train.shape , Y_train.shape)
print(X_test.shape , Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train , Y_train)
```

```python
prediction = reg.predict(X_test)
```

```python
import seaborn as sns
sns.distplot(Y_test-prediction)
```

```python
from sklearn import metrics
```

```python
from sklearn import metrics

import numpy as np

# Mean Absolute Error

print('MAE: ', metrics.mean_absolute_error(Y_test , prediction))

# Mean Squared Error

print('MSE: ', metrics.mean_squared_error(Y_test, prediction))

# Root Mean Squared Error

print('RMSE: ', np.sqrt(metrics.mean_squared_error(Y_test, prediction)))
```
